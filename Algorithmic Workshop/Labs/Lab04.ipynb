### Import

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 1-  Draw Graph

```python
G = {"0": ["1", "3", "8"],
     "1": ["0", "7"],
     "2": ["3", "7", "5"],
     "3": ["0", "4"],
     "4": ["3", "8"],
     "5": ["2", "6"],
     "6": ["5"],
     "7": ["1", "2"]}

graph = nx.Graph(G)
color_map = ['green' if node == "0" else 'yellow' for node in graph]        

nx.draw_spring(graph, node_size=800, node_color=color_map, with_labels=True)
```

---

# 2- DFS path for the Graph

```python
dfs_output = list(nx.dfs_preorder_nodes(graph, source="0"))

print(dfs_output)
```

---

# 3- BFS Path for the graph

```python
bfs_output = list(nx.bfs_tree(graph, source="0"))

print(bfs_output)
```

---

# 4- Dijkstra

```python
G = nx.Graph()

G.add_edge(0,1,weight=8)
G.add_edge(0,2,weight=9)
G.add_edge(1,3,weight=1)
G.add_edge(1,5,weight=7)
G.add_edge(1,6,weight=9)
G.add_edge(2,5,weight=6)
G.add_edge(2,6,weight=4)
G.add_edge(3,7,weight=4)
G.add_edge(4,7,weight=7)
G.add_edge(5,7,weight=2)
G.add_edge(6,7,weight=6)

pos = nx.spring_layout(G)
nx.draw(G,pos, node_size=900)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels, font_weight='bold')
plt.axis("off")
plt.show()
```

```python
distance, path = nx.single_source_dijkstra(G, source=0, target=7)
```

```python
print("Shortest path from 0 to 7 is:  {}".format(path))
print("The weight of this path is: {}".format(distance))
```
